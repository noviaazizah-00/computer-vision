```python
# Load video
cap = cv2.VideoCapture('elsa.mp4')

# iterate for each frame in video
while cap.isOpened():

    # get image on each frame
    ret, frame = cap.read()

    # show image
    cv2.imshow('Frame', frame)

    # wait 25ms per frame and close using 'q'
    if cv2.waitKey(25) == ord('q'):
        break

# close video dan window
cap.release()
cv2.destroyAllWindows()
```

```python
# Load video
cap = cv2.VideoCapture('elsa.mp4')

# iterate setiap frame video
while cap.isOpened():

    # membaca frame
    ret, frame = cap.read()

    if ret == True:

        # menampilkan frame
        cv2.imshow('Frame', frame)

        # tekan q untuk keluar
        if cv2.waitKey(25) == ord('q'):
            break

        # tekan s untuk screenshot
        if cv2.waitKey(25) == ord('s'):
            cv2.imwrite("foto_from_elsa.jpg", frame)
            print("foto captured!")

    else:
        break

# menutup video
cap.release()

# menutup semua window
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
ret, frame = cap.read()

if ret :
    cv2.imwrite("my_photo.jpg", frame)
else :
    print("can't save photo")

# close vidio
cap.release()
```

```python
cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 320)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 240)

# Define the codec and create Videowriter object
# fourcc adalah kode untuk menentukan format video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter('output_2.mp4', fourcc, 20.0, (320, 240))

cv2.VideoWriter
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        # menyimpan frame ke video
        out.write(frame)

        cv2.imshow('frame', frame)
        key = cv2.waitKey(1)

        if key == ord('q'):
            break
        if key == ord('s'):
            cv2.imwrite("foto_from.jpg", frame)
            print("foto captured!")

    else:
        break

# release semua resource
cap.release()
out.release()
cv2.destroyAllWindows()
```
